```python
def get_hand_type(hand):
  cards = hand[0]
  hand_eval = {"A": 0, "K": 0, "Q": 0, "J": 0, "T": 0, "9": 0, "8": 0, "7": 0,
               "6": 0, "5": 0, "4": 0, "3": 0, "2": 0}
  hand_hex = {"A": "E", "K": "D", "Q": "C", "J": "B", "T": "A" }

  hex_text = ""
  for card in cards:
    hand_eval[card] += 1
    if card.isdigit():
      hex_text += card
    else:
      hex_text += hand_hex[card]
  a = list(hand_eval.values())
  a.sort(reverse=True)
  hand_key = int("".join(str(char) for char in a if char > 0))
  total_hand_strenght = int(hex_text, 16)
  return hand_key, (total_hand_strenght, hand[1])
```

```python
def get_hand_type_second(hand):
  cards = hand[0]
  hand_eval = {"A": 0, "K": 0, "Q": 0, "J": 0, "T": 0, "9": 0, "8": 0, "7": 0,
               "6": 0, "5": 0, "4": 0, "3": 0, "2": 0}
  hand_hex = {"A": "E", "K": "D", "Q": "C", "J": "1", "T": "A" }

  hex_text = ""
  count_J = 0
  for card in cards:
    if card == "J":
      count_J += 1
    else:
      hand_eval[card] += 1
    if card.isdigit():
      hex_text += card
    else:
      hex_text += hand_hex[card]
  a = list(hand_eval.values())
  a.sort(reverse=True)
  hand_key = "".join(str(char) for char in a if char > 0)
  start = "0" if len(hand_key) == 0 else hand_key[0]
  rest = "" if len(hand_key) == 1 else hand_key[1:]
  hand_key = int(str(int(start)+count_J)+rest)
  total_hand_strenght = int(hex_text, 16)
  return hand_key, (total_hand_strenght, hand[1])
```

```python
def count_winnings(dictionary):
  strenght_order = list(dictionary.keys())
  strenght_order.reverse()

  sorted_hands = []
  for strenght in strenght_order:
    hands = {}
    for hand in dictionary[strenght]:
      hands[hand[0]] = int(hand[1])
    sorted = list(hands.keys())
    sorted.sort()


    for i in sorted:
      sorted_hands.append(hands[i])

  total = 0
  for i, bet in enumerate(sorted_hands):
    total += bet * (i+1)
  return total
```

```python
def create_dict(hands, function):
  divided_strengths = {
      "five of kind" : [],
      "four of kind" : [],
      "full house" : [],
      "three of kind" : [],
      "two pair" : [],
      "one pair" : [],
      "high card" : []
  }
  x = function

  for hand in hands:
    hand_key, hand = x(hand)
    divided_strengths[hand_type[hand_key]].append(hand)

  return divided_strengths

```
